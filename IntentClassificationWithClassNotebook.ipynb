```python
%%capture

import IntentClassification as ic
import torch

# specify parameters to use with intent classification with BERT
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
path_to_json = './data_full.json'
seed = 5
num_labels_to_sample = 20
padding_max_length = 128
batch_size = 16
optimizer = 'AdamW'
lr = 2e-5
epochs = 4

# create a pipeline for intent classification
pipeline = ic.IntentClassificationWithBert(device, path_to_json, num_labels_to_sample,
                 seed, padding_max_length, batch_size,
                 optimizer, lr)
```

```python
pipeline.train()
```

```python
true_labels, predictions = pipeline.evaluate()
```

```python
pipeline.plot_confusion_matrix(true_labels, predictions)
```

```python
print(pipeline.get_classification_report(true_labels, predictions))
```
